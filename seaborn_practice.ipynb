# Seaborn

*Seaborn is a Python data visualization library with an emphasis on statistical plots. The library is an excellent resource for common regression and distribution plots, but where Seaborn really shines is in its ability to visualize many different features at once.*

## 1. Data Set
[This data set is from UCI](https://archive.ics.uci.edu/ml/datasets/Auto+MPG)

## 2. Preprocessing

```python
import numpy as np
import pandas as pd
%pylab inline
import seaborn as sns; sns.set_style('darkgrid')
# %pylab inline: to make graph be shown
```

```python
# read data from csv
df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")
df.tail()
```

```python
# we need column names
names = ['mpg', 'cylinders', 'displacement','horsepower','weight',\
         'acceleration', 'model_year','origin', 'car_name']

named_df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data",\
                 sep='\s+', names = names)
```

```python
df.tail()
```

```python
# add column 'maker'
named_df['maker'] = named_df.car_name.map(lambda record: record.split()[0])

# chagnge 'origin' to be informative
named_df.origin = named_df.origin.map({1: "America", 2: "Europe", 3: "Asia"})
readible_df = named_df
readible_df.tail()
```

```python
# there are '?' in the dataframe
readible_df.isnull().sum() # it does not count '?'
```

```python
# let's make "?" nan first and drop it
no_nan_df = readible_df.applymap(lambda record: np.nan if record == "?" else record).dropna()
no_nan_df.tail()
```

```python
preprocessed_df = no_nan_df
```

### 3-2. pairplot and PariGrid

```python
want_to_pair_cols = ["mpg", "horsepower", "weight", "origin"]
g = sns.pairplot(preprocessed_df[want_to_pair_cols], hue = 'origin')
```

```python
g = sns.PairGrid(preprocessed_df[want_to_pair_cols], hue = 'origin')
# above diagonal
g.map_upper(sns.regplot)
# below diagonal
g.map_lower(sns.residplot)
# diagonal
g.map_diag(plt.hist)

for ax in g.axes.flat:
    plt.setp(ax.get_xticklabels(), rotation = 45)

g.add_legend()
g.set(alpha = 0.5)
```

### 3-3. jointplot and JointGrid

```python
sns.jointplot('mpg', 'horsepower', data = preprocessed_df, kind = 'kde')
```

```python
sns.jointplot('horsepower', 'mpg', data = preprocessed_df, kind = 'reg')
```

```python
g = sns.JointGrid(x = "horsepower", y = 'mpg', data = preprocessed_df)
g.plot_joint(sns.regplot, order = 2)
g.plot_marginals(sns.distplot)
```
